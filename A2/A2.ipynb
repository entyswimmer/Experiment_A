```python
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib
import numpy as np
import os
```

```python
def plot1(data, sheet):
    df = pd.read_excel(data, sheet_name=sheet)
    x = df[df.columns[0]]
    
    xlabel=input("x軸のラベル名を入力: ")
    ylabel=input("y軸のラベル名を入力: ")
    fig_name=input("グラフの保存名を入力: ")
    output_dir = "./Pictures"
    os.makedirs(output_dir, exist_ok=True)
    output_file = os.path.join(output_dir, fig_name + ".png")
    
    plt.figure(figsize=(8, 5))
    markers = ['o', 's', 'D', '^', 'v', 'x', '+', '*', 'p', 'h']
    for idx, col in enumerate(df.columns[1:]):
        y = df[col]
        marker = markers[idx % len(markers)]
        plt.plot(x, y, marker=marker, label=col)

    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.xlim(-150, 150)
    plt.ylim(-14, 14)
    plt.legend(fontsize=14)
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(output_file)
    plt.show()
```

```python
def tabele(data, sheet):
    df = pd.read_excel(data, sheet_name=sheet)
    latex_table = df.to_latex(index=False)
    print(latex_table)
```

```python
data = "./DataSet/data_A2.xlsx"
```

```python
sheet1 = "amp10"
plot1(data, sheet1)
```
