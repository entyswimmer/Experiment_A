```python
def max_mu(df):
    df = df.sort_values("H")
    dx = df["H"].diff()
    dy = df["B"].diff()
    slope = dy/dx
    max_mu = slope.max()
    idx = slope.idxmax()
    print(f"最大傾き: {slope[idx]}")
    print(f"区間: H={df.loc[idx-1, 'H']} to H={df.loc[idx, 'H']}")
```

```python
max_mu(df_l)
```

```python
max_mu(df_s)
```
