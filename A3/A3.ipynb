```python
#テスツ
def plot(df):
    col_A = df.columns[0]   # A
    col_B = df.columns[1]   # B
    col_D = df.columns[3]   # D
    x = df[col_A]
    y1 = df[col_B]
    y2 = df[col_D]
    plt.figure(figsize=(8, 6))
    plt.plot(x, y1, label="理論値")
    plt.plot(x, y2, label="実測値")
    plt.xlabel("Time (s)", fontsize=14)
    plt.xlim(6, 11)
    plt.ylabel("Speed (rad/s)", fontsize=14)
    plt.legend(loc="upper right", fontsize=12)
    plt.show()
```

```python
def save_graph(DATA_DIR):
    excel_files = glob.glob(os.path.join(DATA_DIR, "*.xlsx"))
    for file in excel_files:
        print(f"\nProcessing: {file}")
        # Excel読み込み
        df = pd.read_excel(file)
        col_A = df.columns[0]   # A
        col_B = df.columns[1]   # B
        col_D = df.columns[3]   # D
        x = df[col_A]
        y1 = df[col_B]
        y2 = df[col_D]
        plt.figure(figsize=(8, 4))
        plt.plot(x, y1, label="理論値", color="navy")
        plt.plot(x, y2, label="実測値", color="crimson")
        plt.xlabel("Time (s)", fontsize=14)
        plt.xlim(14, 21) #場合によって変更
        plt.ylabel("Position (rad)", fontsize=14) #場合によって変更
        plt.legend(loc="upper right", fontsize=12)
        plt.grid(True)
        output_dir = "./Pictures"
        os.makedirs(output_dir, exist_ok=True)
        output_file = os.path.join(output_dir, os.path.basename(file) + ".png")
        plt.tight_layout()
        plt.savefig(output_file)
        plt.close()
        print(f"Saved graph → {output_file}")
```
